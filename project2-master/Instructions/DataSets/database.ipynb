```python
import pandas as pd
import numpy as np
import csv
```

```python
# Create reference to CSV file
csv_path = "../../../DataSets/the-office-lines - scripts.csv"

# Import the CSV into a pandas DataFrame
off_df = pd.read_csv(csv_path, low_memory=False)
off_df1 = off_df.drop(columns=['scene', 'id'])
off_df1['s0e0'] = off_df1.season.astype(str).str.cat(off_df1.episode.astype(str).str.zfill(2))
off_df2 = off_df1[['s0e0', 'line_text', 'speaker', 'deleted', 'season','episode']]
off_df3 = off_df2.set_index('s0e0')
off_df4 = off_df3.drop(columns=['season', 'episode'])
off_df4
```

```python
#off_df4.to_csv('../../DataSets/line_table1.csv')
```

```python
off_df5 = off_df4.groupby(['s0e0','speaker']).count().reset_index().drop(columns=['deleted'])
off_df5.set_index('s0e0').head()
```

```python
#off_df5.to_csv('../../DataSets/line_counts.csv')
```

```python
episode_selection = off_df5.loc[off_df5['s0e0']=='101',:].set_index('speaker').sort_values(by='line_text',ascending=False)
```

```python
episode_selection.plot.bar()
```
